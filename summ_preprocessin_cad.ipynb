```python
import pandas as pd

df = pd.read_table('phenocode-429.3.tsv')
print(df)
```

```python
df_ = df[df["chrom"]==1]
```

```python
print(df_.shape,df.shape)
```

```python
for i in range(df_.shape[0]):
    if df_.loc[i,"beta"]<0:
        df_.loc[i,"Direction"]=-1
        df_.loc[i,"beta"]*=-1
    else:
        df_.loc[i,"Direction"]=1
```

```python
# for i in range(df.shape[0]):
#     df_.loc[i,"Predictor"] = "%d:%d" % (df_.loc[i,"#chrom"],df_.loc[i,"pos"])
```

```python
df_.rename(columns = {'ref' : 'A1'}, inplace = True)
df_.rename(columns = {'alt' : 'A2'}, inplace = True)
df_.rename(columns = {'rsids' : 'Predictor'}, inplace = True)
df_.rename(columns = {'beta' : 'Stat'}, inplace = True)
```

```python
df_["n"] = 407690
```

```python
df_.drop(["chrom","pos","nearest_genes",'consequence',"pval",'sebeta','af','tstat',"ac"],axis=1,inplace=True)
```

```python
df_ = df_[["Predictor","A1","A2","Direction","Stat","n"]]
```

```python
df_.dropna(subset="Predictor", inplace=True)
```

```python
df_.to_csv('t2d.summaries', index=False, header=True, sep=" ")
```
