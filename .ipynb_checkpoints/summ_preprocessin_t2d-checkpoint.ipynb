```python
import pandas as pd

df = pd.read_table('phenocode-250.2.tsv')
print(df)
```

```python
df_ = df[df["chrom"]==1]
```

```python
print(df_.shape,df.shape) 
```

```python
for i in range(df_.shape[0]):
    if df_.loc[i,"beta"]<0:
        df_.loc[i,"Direction"]=-1
        df_.loc[i,"beta"]=df_.loc[i,"beta"]*(-1)
    else:
        df_.loc[i,"Direction"]=1
```

```python
df_.rename(columns = {'ref' : 'A1'}, inplace = True)
df_.rename(columns = {'alt' : 'A2'}, inplace = True)
df_.rename(columns = {'rsids' : 'Predictor'}, inplace = True)
df_.rename(columns = {'beta' : 'Stat'}, inplace = True)
```

```python
df_["n"] = 407690
```

```python
df_.drop(["chrom","pos","nearest_genes",'consequence',"pval",'sebeta','af','tstat',"ac"],axis=1,inplace=True)
```

```python
df_ = df_[["Predictor","A1","A2","Direction","Stat","n"]]
```

```python
df_.dropna(subset="Predictor", inplace=True)
```

## inconsistent genotypes

```python
from pandas_plink import read_plink
(bim, fam, bed) = read_plink("ukb_chr1_v2.bed",verbose=False)
bim
```

```python
df_[df_.duplicated(subset="Predictor",keep=False)]
```

- 원래 duplicated -> 15908 pairs
- 이 중 bim과 겹치고 inconsistent -> 294 -> drop
- still duplicated -> 15614 -> only first

```python
df_idx = df_.reset_index(drop=False, inplace=False)
df_idx
```

```python
df_INNER_JOIN = pd.merge(df_idx, bim, left_on='Predictor', right_on='snp', how='inner')
```

```python
df_INNER_JOIN
```

```python
dup = df_INNER_JOIN[df_INNER_JOIN.duplicated(subset="Predictor",keep=False)]
dup
```

```python
#dup.set_index('index', inplace=True)
inconsis_idx = dup[dup["A2"]!=dup["a1"]].index
df_consis = df_.drop(inconsis_idx)
```

```python
df_consis[df_consis.duplicated(subset="Predictor",keep=False)]
```

```python
df_consis.drop_duplicates(subset='Predictor', inplace=True) #2223733 -> 2197173 
```

```python
#modify the rest inconsistent genotypes
df_consis.loc[df_consis["Predictor"]=='rs199984518','A2'] = bim[bim["snp"]=='rs199984518']['a1'].values[0]
df_consis.loc[df_consis["Predictor"]=='rs60822373','A2'] = bim[bim["snp"]=='rs60822373']['a1'].values[0]
```

```python
df_consis.to_csv('t2d.summaries', index=False, header=True, sep=" ")
```
